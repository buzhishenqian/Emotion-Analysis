```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("hotel_comment.csv",header=None)
df
```

```python
df = df.sample(frac=1)
df
```

```python
dev_percentage=0.15
splite_index=int(dev_percentage*len(df))
len(df),splite_index
```

```python
dev_csv=df[:splite_index]
train_csv=df[splite_index:]
train_csv,dev_csv
```

```python
train_csv.to_csv("train.csv",header=0,index=0)
dev_csv.to_csv("valid.csv",header=0,index=0)
```
